```python
from selenium import webdriver
from selenium.webdriver.common.by import By
# these packages will be used in the final versions
# from selenium.webdriver.common.keys import Keys
# from selenium.webdriver.common.action_chains import ActionChains
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
page_no = 1
driver = webdriver.Chrome()
# last page - 146
```

```python
driver.get("https://etfdb.com/screener/#page=1")
# after driver is initialised, create a for loop iterating through all pages 
```

```python
# new tab for each page
```

```python
# extract first 7 columns of each row
columns = driver.find_elements(By.TAG_NAME, 'th')
headers = []

for i in columns:
    headers.append(i.text.strip())
headers = headers[:7]
```

```python
headers
```

```python
rows = driver.find_elements(By.TAG_NAME, 'tr')

table_data = []

for r in rows:
    cells = r.find_elements(By.TAG_NAME, 'td')

    if cells:
        row_data = []

        for c in cells:
            row_data.append(c.text.strip())
        table_data.append(row_data[:7])
```

```python
table_data = table_data[:-12]
table_data
```

```python
df = pd.DataFrame(table_data, columns=headers)
df.head()
```
